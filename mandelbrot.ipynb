```python
import math
import numpy as np

%matplotlib inline
import matplotlib as mp
import matplotlib.pyplot as plt
mp.rcParams['text.usetex']=True
```

MandelbrotSet
=============
This class collects all the points of the mandelbrot set.

Constructor
-----------
The constructor takes in 8 parameters"

- x_min: minimum x value
- x_max: maximum y value
- y_min: minimum y value
- y_max: maximum y value
- npix_x: number of pixels to be drawn in the x direction
- npix_y: number of pixels to be drawn in the y direction
- max_itr: maximum number of iterations to try when checking if point is in the mandelbrot set (default = 10)
- escape_radius: escape radius used to determine if the point is in or not (default = 2)

```python
class MandelbrotSet(object):
    """
    set of mandelbrot points
    """
    # -----------------------------------------------------------------------------
    def __init__( self
                , x_min
                , x_max
                , y_min
                , y_max
                , npix_x
                , npix_y
                , max_itr = 10
                , escape_radius = 2
                ):
        # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
        print '------------------------------------------------------------------'
        print 'Generating Mandelbrot Set using:'
        print '\tx: [%s, %s] - res: %s' % (x_min, x_max, npix_x)
        print '\ty: [%s, %s] - res: %s' % (y_min, y_max, npix_y)
        print 'maximum iterations: %d' % max_itr

        self.x_min         = x_min
        self.x_max         = x_max
        self.y_min         = y_min
        self.y_max         = y_max
        self.npix_x        = npix_x
        self.npix_y        = npix_y
        self.max_itr       = max_itr
        self.escape_radius = escape_radius

        self.x_points = np.linspace(x_min, x_max, npix_x)
        self.y_points = np.linspace(y_min, y_max, npix_y)
        
        self.x_grid, self.y_grid = np.meshgrid(self.x_points, self.y_points)
        
        self.constructSet()

    # -----------------------------------------------------------------------------
    def constructSet(self):
        # construct the ndarray of complex values
        c = self.x_grid + self.y_grid*1j

        # construct ndarrays to store if the point is in the MS and if not, what is the escape speed
        num_rows = c.shape[1]
        num_cols = c.shape[0]
        self.in_set            = np.zeros(c.shape) + 1
        self.escape_speed      = np.zeros(c.shape)
        self.norm_escape_speed = np.zeros(c.shape)
        self.final_z           = np.zeros(c.shape)

        # now, loop over the rows and columns to determine if the point is in the MS
        for row in xrange(num_rows):
            for col in xrange(num_cols):            
                # check if this point is in the MS
                itr = 0
                escaped = False

                this_z = 0
                this_c = c[col][row]
                while not escaped and itr < self.max_itr:
                    itr += 1

                    this_z *= this_z
                    this_z += this_c

                    if abs(this_z) > self.escape_radius:
                        escaped = True

                if escaped:
                    self.in_set[col][row] = 0
                    self.escape_speed[col][row] = itr
                    self.norm_escape_speed[col][row] = ( itr
                                                       - (math.log( math.log(abs(this_z))
                                                                  / math.log(self.escape_radius)
                                                                  )
                                                         / math.log(2)
                                                         )
                                                       )
                    self.final_z[col][row] = self.escape_radius
                else:
                    self.final_z[col][row] = abs(this_z)
```

```python
mbs = MandelbrotSet( x_min         = -3.0
                   , x_max         = +1.8
                   , y_min         = -1.5
                   , y_max         = +1.5
                   , npix_x        = 800
                   , npix_y        = 600
                   , max_itr       = 50
                   , escape_radius = 200
                   )
```

```python
plt.plot()
plt.pcolor(mbs.x_grid, mbs.y_grid, mbs.in_set, cmap=plt.cm.gist_earth)
plt.axis('off')
plt.savefig('mandelbrot_in_set.png')
```

```python
plt.plot()
plt.pcolor(mbs.x_grid, mbs.y_grid, mbs.escape_speed, cmap=plt.cm.gist_earth)
plt.axis('off')
plt.savefig('mandelbrot_escape_speed.png')
```

```python
plt.plot()
plt.pcolor(mbs.x_grid, mbs.y_grid, mbs.norm_escape_speed, cmap=plt.cm.gist_earth)
plt.axis('off')
plt.savefig('mandelbrot_norm_escape_speed.png')
```

```python
plt.plot()
plt.pcolor(mbs.x_grid, mbs.y_grid, mbs.final_z, cmap=plt.cm.gist_earth, vmax=2)
plt.axis('off')
plt.savefig('mandelbrot_final_z.png')
```
